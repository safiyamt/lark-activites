```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Dataset_w6.xlsx - Data.csv')
df
```

```python
data=df.groupby('Region').agg({'Surface area (km2)':'sum','Population in thousands (2017)':'sum'})
```

```python
plt.figure(figsize=(8,12))
labels=df['Region'].value_counts().index

plt.pie(data['Population in thousands (2017)'],labels=labels,autopct='%1.1f%%')
plt.title('population in thousands (2017)')
plt.show()
```

```python
data2=data.sort_values(['Population in thousands (2017)'],ascending=False)
plt.figure(figsize=(8,12))
data2.plot(kind='bar')
```

```python
data3=df.sort_values(['GDP per capita (current US$)'],ascending=False).head(10)
labels=data3['country'].value_counts().index
plt.pie(data3['GDP per capita (current US$)'],labels=labels,autopct='%1.1f%%')
plt.title('GDP per capita (current US$) in percentage')
plt.show()
```

# 4. On average which region has the highest GDP per capita (current US$)

```python
df1=df.groupby('Region')['GDP per capita (current US$)'].agg(['mean'])
df1[df1['mean']==max(df1['mean'])]
```

# On average which region has the highest International trade: Imports (million US$)

```python
df2=df.groupby('Region')['International trade: Imports (million US$)'].agg(['mean'])
df2[df2['mean']==max(df2['mean'])]

```

# Find out which region has more consistent with respect to GDP per capita (current US$)

```python
df3=df.groupby('Region')['GDP per capita (current US$)'].agg(['std'])
df3[df3['std']==min(df3['std'])]
```
